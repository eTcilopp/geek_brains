```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/6183122/attachment/33af8537bca92e8880cabb6af8d3446b.csv')
```

```python
data.head(2)
```

```python
from matplotlib import pyplot as plt
data.plot(kind='scatter', x='residual sugar', y='chlorides', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.histplot(data=data, x='quality', kde=True)
```

```python
sns.barplot(x=data.quality, y=data.alcohol, palette="Blues_d", hue=data.quality, legend=False)
```

```python
sns.histplot(data=data.query('quality > 6'), x='residual sugar', kde=True)
```

```python
sns.histplot(data=data.query('quality < 6'), x='residual sugar', kde=True)
```
